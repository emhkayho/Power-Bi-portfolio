#### Let us deal with accented names so our data can be uniform

```python
#import the unidecode library to convert accented characters to plain english alternatives
#identify the columns that have accented characters using the 'dtypes' attribute of the dataframe
from unidecode import unidecode
object_cols = geography.select_dtypes(include=['object']).columns
```

```python
#we then replace accented charaters with plain english alternatives using unidecode function
for geolocation in object_cols:
    geography[geolocation] = geography[geolocation].apply(lambda x: unidecode(x))
```

```python
geography.head(8)
```

We see that the names in the 'geolocation_city' attributes are now uniform

we see that some of the comment message and comment title are nan, well would be filling them as None where there are null.

Now let us deal with the datetime attributes

```python
#Split the datetime column into date and time
reviews[['review_date_creation', 'Creation_Time']] = reviews['review_creation_date'].str.split(' ', expand = True)

#Convert the date column to datetime datatype
reviews['review_date_creation'] = pd.to_datetime(reviews['review_date_creation'])

reviews.drop('review_creation_date', axis = 1, inplace = True)

#updated dataset
reviews.head()
```

```python
#Split the datetime column into date and time
reviews[['answer_timestamp', 'answer_Time']] = reviews['review_answer_timestamp'].str.split(' ', expand = True)

#Convert the date column to datetime datatype
reviews['answer_timestamp'] = pd.to_datetime(reviews['answer_timestamp'])

reviews.drop('review_answer_timestamp', axis = 1, inplace = True)

#updated dataset
reviews.head()
```

```python
object_cols = reviews.select_dtypes(include=['object']).columns

#we then replace accented charaters with plain english alternatives using unidecode function
for comment_message in object_cols:
    reviews[comment_message] =reviews[comment_message].apply(lambda x: unidecode(x))
    
reviews.head()
```

### Looking at the orders dataset

```python
orders.head()
```

```python
#Overview
orders.info()
```

```python
datetime_cols = ['order_purchase_timestamp', 'order_approved_at', 'order_delivered_carrier_date',
                 'order_delivered_customer_date', 'order_estimated_delivery_date']

for col in datetime_cols:
    #split the datetime column into date and time
    orders[[col+'_Date', col+ '_Time']] = orders[col].str.split(' ', expand = True)
    #Convert the date column to datetime datatype
    orders[col+ '_Date'] = pd.to_datetime(orders[col+'_Date'])
    #Taking only the date into a new datetime column
    orders[col+ '_NewDate'] = pd.to_datetime(orders[col+'_Date'].dt.date)
    #drop the original datetime date and, time columns
    orders.drop([col,col+'_Date', col+'_Time'], axis = 1, inplace = True)
```

```python
orders.info()
```

```python
#filling null date in 5, 6, 4
#We would fill the data mainly with potential estimates
orders['order_approved_at_NewDate'] = orders['order_approved_at_NewDate'].fillna(orders['order_purchase_timestamp_NewDate'])
orders.info()
```

```python
orders['order_delivered_carrier_date_NewDate'] = orders['order_delivered_carrier_date_NewDate'].fillna(orders['order_approved_at_NewDate'])
orders.info()
```

```python
#WE use the estimated delivery time to fill the nan
orders['order_delivered_customer_date_NewDate'] = orders['order_delivered_customer_date_NewDate'].fillna(orders['order_estimated_delivery_date_NewDate'])
orders.info()
```
